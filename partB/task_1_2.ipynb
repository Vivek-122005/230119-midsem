# Task 1.2 - Key Assumptions of the Paper

**Paper:** *Early Prediction on Time Series: A Nearest Neighbor Approach*

The method's validity depends on assumptions that are specific to prefix-based nearest-neighbor early classification (not generic ML assumptions).

---

## Assumption 1: Early Prefixes Already Contain Class-Discriminative Information

**Assumption:** For a meaningful subset of samples, class-separating signal appears before full sequence length.

**Why the method needs it:** If early prefixes are informative, MPL can be small and prediction can be early. If not, ECTS either predicts unreliably or must defer to late/full length.

**Violation scenario:** Two classes are almost identical for the first 70-80% of timestamps and diverge only near the end. In this case, early decisions become close to random.

**Paper reference:** Motivation and early classification formulation (Sections 1-3).

---

## Assumption 2: Distance-Based Nearest Neighbors Are Label-Meaningful

**Assumption:** The chosen distance in prefix/full spaces retrieves neighbors that are genuinely similar for class prediction.

**Why the method needs it:** Both the full-length baseline and early-prefix decisions are 1NN-based. If nearest neighbors are poor proxies for label similarity, MPL and early decisions degrade together.

**Violation scenario:** Temporal phase shifts or heavy noise make Euclidean closeness mismatch true class semantics.

**Paper reference:** 1NN method foundation and Euclidean-distance experiments (Sections 3 and 5).

---

## Assumption 3: Decision Stability Improves with Increasing Prefix Length

**Assumption:** As `l` increases, predicted neighbor label behavior tends to stabilize.

**Why the method needs it:** MPL requires an earliest reliable point. Without stabilization, there is no trustworthy early stopping boundary.

**Violation scenario:** Highly ambiguous/noisy sequences cause repeated label flips across many prefix lengths.

**Paper reference:** Serial classifier concept, MPL definitions, and consistency constraints (Sections 2-4).

---

## Assumption 4: Training Data Adequately Covers Test-Time Pattern Diversity

**Assumption:** Training examples are representative enough for meaningful neighbor retrieval and MPL estimation.

**Why the method needs it:** ECTS learns timing and discrimination structure from training samples. Weak coverage produces unreliable neighbors and unstable early decisions.

**Violation scenario:** Domain shift introduces unseen dynamics in test data; both early and full 1NN quality drop.

**Paper reference:** Support/cluster-based training logic and benchmark-driven empirical setup (Sections 4-5).

---

## Why These Assumptions Matter for Later Tasks

Task 3 failure-mode analysis is strongest when it directly breaks one of these assumptions. In this project, the synthetic failure mode intentionally violates Assumption 1 (uninformative early prefixes), which creates poor short-prefix accuracy and late recovery.
